```python
# Load all sheets of the Excel file
all_sheets = pd.read_excel('/kaggle/input/stock-exchange-data-2023/Bucharest Stock Exchange Data - YTD 2023.xlsx', sheet_name=None)

# Display the names of the sheets
all_sheets.keys()
```

```python
# Display the first few rows of each sheet
for sheet_name, df in all_sheets.items():
    print(f"First few rows of the '{sheet_name}' sheet:")
    display(df.head())
    print("\n")
```

```python
# Define a function to summarize the data quality
def summarize_data(df):
    summary = pd.DataFrame(df.dtypes, columns=['Data Type'])
    summary['Missing Values'] = df.isnull().sum()
    summary['Unique Values'] = df.nunique()
    return summary

# Summarize the data in each sheet
data_summary = {}
for sheet_name, df in all_sheets.items():
    print(f"Data Summary for '{sheet_name}' Sheet:")
    data_summary[sheet_name] = summarize_data(df)
    display(data_summary[sheet_name])
    print("\n")
```
